```
# regex for removing punctuation!
import re
# pickle
import pickle

#Pre-cleaning text
with open('friends.pickle', 'rb') as handle:
    episodes_script = pickle.load(handle)

for i in range(len(episodes_script)):
  # Line breaks
  episodes_script[i] = re.sub(r"\n"," ",episodes_script[i])
  # Scriptwriter's notes
  episodes_script[i] = re.sub(r"\(.*?\) ","",episodes_script[i])
  episodes_script[i] = re.sub(r"\[.*?\] ","",episodes_script[i])
  # Character identifier
  episodes_script[i] = re.sub(r"\w*?: ","",episodes_script[i])
  # Extra spaces
  episodes_script[i] = re.sub(r"\s\s+"," ",episodes_script[i])
  # Punctuation
  episodes_script[i] = re.sub(r"[.,?!]","",episodes_script[i])
  

with open('friends_clean.pickle', 'wb') as handle:
    pickle.dump(episodes_script, handle, protocol=pickle.HIGHEST_PROTOCOL)

```

```
# nltk preprocessing magic
import nltk
from nltk.tokenize import word_tokenize
# nltk.download('punkt')
# nltk.download('stopwords')
from nltk.corpus import stopwords
from collections import Counter
import pickle
import re

with open('friends_clean.pickle', 'rb') as handle:
    friends = pickle.load(handle)

friends_tokenized = []
for episode in friends:
  friends_tokenized.append(word_tokenize(episode))

#Fixing 'ca', 'wo', 'n't'
for episode in friends_tokenized:
  for j in range(len(episode)):
    if episode[j] == "ca":
      episode[j] == "can"
    if episode[j] == "wo":
      episode[j] == "will"
    if episode[j] == "n't":
      episode[j] == "not"

```

```
stop_words = stopwords.words('english')
episodes = []
episodes_bow = []
for episode in friends_tokenized:
  episodes.append([word for word in episode if word not in stop_words])
for episode in episodes:
  bow = Counter(episode)
  episodes_bow.append(bow)

```
