```
from bs4 import BeautifulSoup as bs
import requests
import pickle

episodes_list = ["https://crazy-internet-people.com/site/gilmoregirls/pages/s1/s1s/" + str(i) for i in list(range(1,22))] + \
         ["https://crazy-internet-people.com/site/gilmoregirls/pages/s2/s2s/" + str(i) for i in list(range(22,44))] + \
         ["https://crazy-internet-people.com/site/gilmoregirls/pages/s3/s3s/" + str(i) for i in list(range(44,66))] + \
         ["https://crazy-internet-people.com/site/gilmoregirls/pages/s4/s4s/" + str(i) for i in list(range(66,88))] + \
         ["https://crazy-internet-people.com/site/gilmoregirls/pages/s5/s5s/" + str(i) for i in list(range(88,110))] + \
         ["https://crazy-internet-people.com/site/gilmoregirls/pages/s6/s6s/" + str(i) for i in list(range(110,132))] + \
         ["https://crazy-internet-people.com/site/gilmoregirls/pages/s7/s7s/" + str(i) for i in list(range(132,153))]


episode_scripts = []  
for episode in episodes_list:
  print(episode)
  html = requests.get(f"{episode}.html")
  soup = bs(html.content)
  search = soup.find_all("p") 
  episode_scripts.append(search[0].text)

with open('gg.pickle', 'wb') as handle:
    pickle.dump(episode_scripts, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open('gg.txt', 'w') as f:
        f.write("%s\n" % episode_scripts)
```
