```
from bs4 import BeautifulSoup as bs
import requests
import pickle

episodes_list = ["010" + str(i) if len(str(i))==1 else "01" + str(i) for i in list(range(1,25))] + \
         ["020" + str(i) if len(str(i))==1 else "02" + str(i) for i in list(range(1,25))] + \
         ["030" + str(i) if len(str(i))==1 else "03" + str(i) for i in list(range(1,26))] + \
         ["040" + str(i) if len(str(i))==1 else "04" + str(i) for i in list(range(1,24))] + \
         ["050" + str(i) if len(str(i))==1 else "05" + str(i) for i in list(range(1,24))] + \
         ["060" + str(i) if len(str(i))==1 else "06" + str(i) for i in list(range(1,25))] + \
         ["070" + str(i) if len(str(i))==1 else "07" + str(i) for i in list(range(1,25))] + \
         ["080" + str(i) if len(str(i))==1 else "08" + str(i) for i in list(range(1,24))] + \
         ["090" + str(i) if len(str(i))==1 else "09" + str(i) for i in list(range(1,24))] + \
         ["100" + str(i) if len(str(i))==1 else "10" + str(i) for i in list(range(1,17))] + \
         ["1017-1018"]


episode_scripts = []  
for episode in episodes_list:
  print(episode)
  html = requests.get(f"https://fangj.github.io/friends/season/{episode}.html")
  soup = bs(html.content)
  search = soup.find_all("p")
  target_string = ""
  for element in search[1:]:
    target_string += " " + element.text 
  episode_scripts.append(target_string)


with open('friends.pickle', 'wb') as handle:
    pickle.dump(episode_scripts, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open('friends.txt', 'w') as f:
        f.write("%s\n" % episode_scripts)
```
